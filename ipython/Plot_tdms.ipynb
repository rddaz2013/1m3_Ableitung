# Daten transformieren von csv in tdms

```
from nptdms import TdmsWriter, RootObject, GroupObject, ChannelObject, TdmsFile 

import numpy as np
import pandas as pd
import os
```

```
base_path="/content/1m3_Ableitung/data/20L_Example"
file_names=os.listdir(base_path)
```

```
file_names
```

```
tdms_file = TdmsFile(base_path+'/20L_Data.tdms')
```

```
root_object = tdms_file.object()

# Iterate over all items in the properties dictionary and print them
for name, value in root_object.properties.items():
    print("{0}: {1}".format(name, value))

# Get a single property value
#property_value = root_object.property("my_property_name")
```

```
Messung = []

for group in tdms_file.groups():
  for channel in tdms_file.group_channels(group):
        # Access dictionary of properties:
        properties = channel.properties
        Messung.append([properties['Messung'],channel.channel])
        # Access numpy array of data for channel:
        # data = channel.data
        # do stuff with data and properties...

#Messung
```

```
time_name ='time [ms]'
data_name ='channel 1 [bar]'
data_time = tdms_file.channel_data('500',time_name)
data_druck = tdms_file.channel_data('500',data_name)
```

```
import altair as alt

data = pd.DataFrame({'x': data_time,
                     'y': data_druck})

alt.Chart(data).mark_point().encode(
    x='x:Q',
    y='y:Q'
)
```
